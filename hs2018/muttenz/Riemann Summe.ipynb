```python
from pylab import *
from math import pi
```

```python
def riemannSum(f, a, b, n):
    if n <= 0:
        print("Error: n must be positive")
        return False
    dx = (double(b)- double(a))/double(n)
    X = arange(a, b+dx, dx)
    leftSum = 0.0
    for x in X:
        leftSum += dx*f(x)
    return leftSum

def riemannplot(f,a,b,n):
    if n<=0:
        print("Error: n must be positive")
        return False
    smoothh= (b-a)/100.0
    X = arange(a, b+smoothh, smoothh)
    plot(X, f(X))
    
    dx = (double(b)-double(a))/double(n)
    riemannX = arange(a, b, dx)
    riemannY = f(riemannX)
    bar(riemannX, riemannY, alpha=0.5, width=dx, facecolor="orange", edgecolor="black", align="edge")
    xlabel('x')
    ylabel('f(x)')
    title("Linke Riemannsumme für f(x) = {0:.5f}".format(riemannSum(f, a, b, n)))
    xlim(a, b)
    show()
```

```python
def f(x):
    return x*sin(x)
```

```python
a = 0
b = 2*pi
n = 20

riemannplot(f, a, b, n)
def F(x):
    return sin(x) - x*cos(x)

leftSum = riemannSum(f, a, b, n)
actualArea = F(b) - F(a)
print("∆A = {0:.3E}\t({1:.3E}%)".format(leftSum - actualArea, 100.0*(leftSum-actualArea)/actualArea))
```
