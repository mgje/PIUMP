```python
%matplotlib inline
import matplotlib as mpl
from matplotlib.pyplot import *
import numpy as np

# Plot Einstellungen
mpl.rc('text', usetex=False)
mpl.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica'],'size' : '16'})
mpl.rc('figure',**{'dpi':160})
#mpl.rc('font',**{'family':'serif','serif':['Palatino']})



def f1(t):
     return t**2*np.exp(-t**2)
def f2(t):
     return t**2*f1(t)

figure(1, figsize = (10, 7))  # Zwei Plots
t = np.linspace(0, 3, 51)
y1 = f1(t)
y2 = f2(t)

subplot(2, 1, 1)
plot(t, y1, 'r-', t, y2, 'bo')
ylabel('y')
axis([t[0], t[-1], min(y2)-0.05, max(y2)+0.5])
legend(['t^2*exp(-t^2)', 't^4*exp(-t^2)'])
title('Obere Abblildung')

subplot(2, 1, 2)
t3 = t[::4]
y3 = f2(t3)

plot(t, y1, 'm--', t3, y3, 'kD')
xlabel('t')
ylabel('y')
axis([0, 4, -0.2, 0.6])
legend(['t^2*exp(-t^2)', 't^4*exp(-t^2)'])
title('Untere Abblidung')

subplots_adjust(hspace=0.5)

show()
```

```python
%matplotlib inline
from matplotlib.pyplot import *
import numpy as np
import operator
plt.rcParams['figure.dpi'] = 160

def Nv(x):
    r = np.where(x < 0, 0.0, x)
    condition = operator.and_(0 <= x, x < 1) 
    r = np.where(condition, x, r)
    condition = operator.and_(1 <= x, x < 2) 
    r = np.where(condition, 2-x, r)
    r = np.where(x >= 2, 0.0, r)
    return r


x = np.linspace(-1, 3, 10)
x2 = np.linspace(-1, 3, 200)
plot(x, Nv(x), 'ro')
plot(x2, Nv(x2), 'b')
legend(['10 points', '200 points'])
axis([x[0], x[-1], -0.1, 1.1])
show()
```

```python
#### %matplotlib inline
import matplotlib as mpl
from matplotlib.pyplot import *
from math import *
import numpy as np


# Plot Einstellungen
mpl.rc('text', usetex=False)
#mpl.rc('font',**{'family':'sans-serif',
#                 'sans-serif':['Helvetica'],'size' : '16'})
mpl.rc('figure',**{'dpi':100})


def anim(tmax, dt, x, function, ymin, ymax, t0=0,\
              xlab='x', ylab='y', hardcopy_stem='tmp_'):
    t = t0
    counter = 0
    while t <= tmax:
        figure()
        y = function(x, t)
        plot(x,y)
        axis([x[0], x[-1], ymin, ymax])
        ylabel(ylab)
        xlabel(xlab)
        title('time=%.1f s' % t)
        #savefig(hardcopy_stem + '%04d.pdf' % counter)
        counter = counter + 1
        t += dt
        

def T(z, t, b=4.5):
      return np.exp(-b*z)*np.cos(t - b*z) 

b = float(2)
n = 401
z = np.linspace(0, 1, n)
t0 = 0.001
TMax = 2*pi
dT = 0.25*pi
yMin = -1.2
yMax = 1.2

anim(TMax, dT, z, T, yMin, yMax, t0, 'z', 'T')
```

```python
#Animation

%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import time
plt.rcParams['figure.dpi'] = 140

#Funktion definieren
def f(x, m, s):
      return (1.0/(np.sqrt(2*pi)*s))*np.exp(-0.5*((x-m)/s)**2)

def pltf(ax,k):
    x = np.linspace(-10, 10, 300)
    y = f(x,0,2/(k+1))
    if ax.lines:
        for line in ax.lines:
            line.set_xdata(x)
            line.set_ydata(y)
    else:
        ax.plot(x, y, "b-")
    fig.canvas.draw()

fig,ax = plt.subplots(1,1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(-3,3)
ax.set_ylim(0,5)
for k in range(20):
    pltf(ax,k)
    time.sleep(1)
```

##  Gruppe $\mathbb{Z}_{10} *$

Die Menge ganze Zahlen10 = {0, ..., 9} bildet mit der Addition modulo 10 als Verknpfung eine Gruppe. Mit der Multiplikation modulo 10 als Verknpfung bildet ganze Zahlen $\mathbb{Z}_{10}$ jedoch keine Gruppe, auch nicht, wenn die 0 ausgenommen wird, die bekanntlich kein inverses Element hat. Es stellt sich heraus, dass noch mehr Zahlen ausgenommen werden mssen, die ebenfalls kein inverses Element haben, nmlich 2, 4, 5, 6 und 8. Dieses sind genau die Zahlen, die einen echten Teiler mit 10 gemeinsam haben. Die schlielich verbleibenden vier Elemente bilden die multiplikative Gruppe ganze Zahlen10*.


Ist die Faktorgruppe $\mathbb{Z}_{10}$ / Nullteiler

2,4,5,8 sind Nullteiler

$[1],[3],[7],[9]$ sind Elemente der Gruppe

```python
M = 10
min_val, max_val = 1, M
ind_array = np.arange(min_val , max_val , 1.0)
x, y = np.meshgrid(ind_array, ind_array)

ymap = [[(x*y)%M for x in ind_array ] for y in ind_array ]

fig = figure(figsize=(12, 8))
xticks(np.arange(0,M,1),np.arange(1,M,1),size=16)
yticks(np.arange(0,M,1),np.arange(1,M,1),size=16)
imshow(ymap,interpolation="nearest",vmin=0,vmax=M,cmap='plasma');
cbar=colorbar()
cbar.ax.tick_params(labelsize=16) 
grid(False)


x, y = np.meshgrid(ind_array-1, ind_array-1)

for i, (x_val, y_val) in enumerate(zip(x.flatten(), y.flatten())):
    c = str(((x_val+1)*(y_val+1))%M) 
    text(x_val, y_val, c, va='center', ha='center')
show()
```

```python
L = [1,3,7,9]
M = 10

ymap = [[(x*y)%M for x in L ] for y in L ]

fig = figure(figsize=(12, 8))
xticks(np.arange(0,5,1),L,size=16)
yticks(np.arange(0,5,1),L,size=16)
imshow(ymap,interpolation="nearest",vmin=0,vmax=M,cmap='plasma');
cbar=colorbar()
cbar.ax.tick_params(labelsize=16) 
grid(False)


x, y = np.meshgrid(np.array(L), np.array(L))

for i, (x_val, y_val) in enumerate(zip(x.flatten(), y.flatten())):
    c = str((x_val*y_val)%M) 
    text(i%4, i//4, c, va='center', ha='center')
show()
```
