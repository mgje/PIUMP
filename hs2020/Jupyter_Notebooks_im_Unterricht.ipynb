```
from math import *

def isPrim(x):
  if x == 2: return True
  Z = floor(sqrt(x))+1
  prim = True
  for z in range(2,Z):
    if x%z == 0:
      prim = False
  return prim

N =  400#@param {type: "number"}
L = []
for x in range(2,N+1):
  if isPrim(x):
    L.append(x)
    
print(L)
```

## Maximum in einer Liste suchen

```
# Übung 3.6 Suche das Maximum
def max(L): # Funktion die das max sucht
    max_el = L[0] # Erstes Element nehmen
    for x in L:
        if x > max_el:
            max_el = x
            
    return max_el

# Liste von Zahlen
L=[5, -3, 37, -10, 7, -16, -18, 12, 15, 11, -5, 
   -26, -27, 25, 29, -28, -21, 22, -26, 35, 22]
print(max(L))
```

## Wurzelberechnung mit Heron

```
zahl = 7 #@param {type: "number"}
```

```
# Übung 4.1 Heronverfahren

from math import sqrt

def heron(x,epsilon=0.001):       # Eine Funktion von x und falls angeben epsilon
    a = x                           # Start mit a = x man kann auch a = 1 wählen
    while (a**2-x)**2 > epsilon**2: # Die Bedingung schaut ob die 
                                    # Abweichung zwischen dem Quadrat der Wurzel (a) und 
                                    # der Fläche(x) kleiner epsilon ist
        a = (a + x/a)/2   # Neuer a Wert ist der Mittelwert vom alten a-Wert und x/a 
        ##print(a)        # Interationsschritte ausgeben
    return a


print ("Heron Berechnung: \t \t Wurzel %i ist %.9f" % 
       (zahl,heron(zahl))) # Heronberechnung
print ("Heron Berechnung genauer: \t Wurzel %i ist %.9f" % 
       (zahl,heron(zahl, 0.0000001))) # Heronberechnung 
                                        # mit kleinerem epsilon
print ("Python Bibliothek: \t \t Wurzel %i ist %.9f" % 
       (zahl,sqrt(zahl)))  # Vergleich mit Bibliothek
```

```
# Übung 4.8 Bisektionsverfahren

def poly_f(x):
    return x**5 - 9*x**4 + 33*x**3 - 60*x**2 + 53*x - 19

eps = 0.00000001 # kleiner Wert

def bisekt(f,a,b):
    if f(a)*f(b)>0:
        print("kein Vorzeichenwechsel in dem Bereich")
        print("es könnte keine Lösung geben!")
        print("Wählen Sie den Bereich grösser")
        return
    while b-a>eps:
        m = (a+b)/2
        if f(m)*f(a)<=0: # Vorzeichenwechsel zwischen a und m
            b = m        # neues Intervall a..m
        if f(m)*f(b)<=0: # Vorzeichenwechsel zwischen m und b
            a = m        # neues Intervall m..b
    return (a+b)/2

x0 = bisekt(poly_f,-10.0,20.0)
print("Die untersuchte Funktion f(x) hat eine Nullstelle bei x=%f \n\
also f(%f) = %.4f"%(x0,x0,poly_f(x0)))
```

## Pi abschätzen mit Monte Carlo



```
from random import *
from math import pi
# Number of darts that land inside.
inside = 0
# Total number of darts to throw.
total = 1000000

# Iterate for the number of darts.
for i in range(0, total):
  # Generate random x, y in [0, 1].
  x2 = random()**2
  y2 = random()**2
  # Increment if inside unit circle.
  if(x2 + y2 < 1.0):
    inside += 1

# inside / total = pi / 4
pi_m = (float(inside) / total) * 4

# It works!
print(pi_m)
print(pi)
```

# Apfelmänchen in der Madelbrot Ebene

```
""" The `maxiter` gives the precision of the computation. `maxiter=200` 
shouldtake a few seconds on most modern laptops."""

# Libraries
import numpy as np
import time
import matplotlib
from matplotlib import colors
import matplotlib.pyplot as plt
```

```
# (xmin,xmax,ymin,ymax)coordinates of the rectangle
# (xn, yn) points to be calculated
# (maxiter) max number of iterations
# (horizon) Radius der beachtet wird

def mandelbrot_set(xmin, xmax, ymin, ymax, xn, yn, maxiter, horizon=2.0):
    X = np.linspace(xmin, xmax, xn, dtype=np.float32)
    Y = np.linspace(ymin, ymax, yn, dtype=np.float32)
    # C 2D-Array of coordinates in the c-plane 
    C = X + Y[:, None]*1j
    # N zero matrix
    N = np.zeros(C.shape, dtype=int)
    Z = np.zeros(C.shape, np.complex64)
    for n in range(maxiter):
        # I is a logic 2d-array
        I = np.less(abs(Z), horizon)
        N[I] = n
        # Matrix operation
        Z[I] = Z[I]**2 + C[I]
    # fields with maxiter will be set to 0
    N[N == maxiter-1] = 0
    return Z, N


```

```
# Default Bereich
xmin, xmax, xn = -2.25, +0.75, 3000/2    #x-grid and dimension
ymin, ymax, yn = -1.25, +1.25, 2500/2    #y-grid and dimension

# Verschiedene Bereiche
# Der letzte unkommentierte wird gezeichnet
xmin, xmax = -1.0 , 0.0
ymin, ymax = -.5  , 0.5

xmin, xmax = -1.0 , -.06  
ymin, ymax = -.5  , -.05

xmin, xmax = -.9 , -.5  
ymin, ymax = -.5  , -.3

xmin, xmax = -.8 , -.5  
ymin, ymax = -.4  , -.1

#xmin, xmax = -1.0 , -.7  
#ymin, ymax = -.5  , -.2

#xmin, xmax = -1.0 , -.6  
#ymin, ymax = -.4  , -.0

#xmin, xmax = -.9 , -.7  
#ymin, ymax = -.3  , -.1

#xmin, xmax = -.85 , -.75  
#ymin, ymax = -.20  , -.10

#xmin, xmax = -.825 , -.775  
#ymin, ymax = -.175  , -.125

xn, yn = 1200,1200

maxiter = 200
horizon = 2.0 ** 40    
log_horizon = np.log(np.log(horizon))/np.log(2)    
Z, N = mandelbrot_set(xmin, xmax, ymin, ymax, int(xn), int(yn), maxiter, horizon)     # Normalized recount as explained in:    
# https://linas.org/art-gallery/escape/smooth.html    
# https://www.ibm.com/developerworks/community/blogs/jfp/entry/My_Christmas_Gift     
# This line will generate warnings for null values but it is faster to    
# process them afterwards using the nan_to_num    
with np.errstate(invalid='ignore'):        
  M = np.nan_to_num(N + 1 - np.log(np.log(abs(Z)))/np.log(2) + log_horizon)     
  dpi = 120    
  width = 10    
  height = 10*yn/xn    
  
  fig = plt.figure(figsize=(width, height), dpi=dpi)    
  ax = fig.add_axes([0.0, 0.0, 1.0, 1.0], frameon=False, aspect=1)     # Shaded rendering    
  light = colors.LightSource(azdeg=315, altdeg=10)    
  M = light.shade(M, cmap=plt.cm.hot, vert_exag=1.5,
                  norm=colors.PowerNorm(0.3), blend_mode='hsv')    
  
  plt.imshow(M, extent=[xmin, xmax, ymin, ymax], interpolation="bicubic")    
  ax.set_xticks([])    
  ax.set_yticks([])     # Some advertisement for matplotlib    
  year = time.strftime("%Y")    
  major, minor, micro = matplotlib.__version__.split('.', 2)    
  #text = ("The Mandelbrot fractal set\n"            
  #        "Rendered with matplotlib %s.%s, %s - http://matplotlib.org" % (major, minor, year))    
  #ax.text(xmin+.025, ymin+.025, text, color="white", fontsize=12, alpha=0.5)     
  plt.show()
```
